### Setting up your Google API Key in Colab

```python
# Import the Python SDK for Google Generative AI
import google.generativeai as genai
# Used to securely store your API key
from google.colab import userdata

# Retrieve the API key from Colab secrets
GOOGLE_API_KEY = userdata.get('GOOGLE_API_KEY')

# Configure the Generative AI library with your API key
genai.configure(api_key=GOOGLE_API_KEY)

# The line below caused an AttributeError because google.generativeai does not have a 'Client' class.
# Removed: client = genai.Client(api_key=os.getenv("GOOGLE_API_KEY"))
```

```python
# The 'google.generativeai' library is already imported and configured in cell eaa9c59b.
# Make sure to run cell eaa9c59b after setting your API key in Colab secrets.

# Instantiate the generative model directly after genai.configure()
# Using 'gemini-3.5-flash' model
model = genai.GenerativeModel("gemini-3.1-flash-lite")

# Generate content using the 'model' object
response = model.generate_content(
    contents=[
     #   "System: You are a helpful assistant specialized in providing information about BellaVista Italian Restaurant.",
      #  "User: What's on the menu?",
       # "Assistant: BellaVista offers a variety of Italian dishes including pasta, pizza, and seafood.",
       # "User: Do you have vegan options?"
 "What is the format of  Contents of  model.generate_content of genai.generartivemodel"
    ]
)

# Print the text content of the response
print(response.text)
```

```python
# The 'google.generativeai' library is already imported and configured in cell eaa9c59b.
# Make sure to run cell eaa9c59b after setting your API key in Colab secrets.

# Instantiate a GenerativeModel for chat
model = genai.GenerativeModel("gemini-3.5-flash")

# Pass initial history using the start_chat method of the model
chat = model.start_chat(
    history=[
        {"role": "user", "parts": ["Hello"]},
        {"role": "model", "parts": ["Great to meet you. What would you like to know?"]},
    ]
)

response = chat.send_message(content="I have 2 dogs in my house.")
print(response.text)

response = chat.send_message(content="How many paws are in my house?")
print(response.text)
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
llm = ChatGoogleGenerativeAI(    model="gemini-3.1-flash-lite",   temperature=0.76, google_api_key=GOOGLE_API_KEY)
response = llm.invoke("Tell me a sexy joke")

print(response.content)
```

```python
from langchain_core.messages import HumanMessage, AIMessage, SystemMessage

messages = [
    SystemMessage(content="You are a helpful AI assistant that provides concise answers."),
    HumanMessage(content="What is the capital of France?"),
    AIMessage(content="The capital of Bangladesh  is Dhaka."),
    HumanMessage(content="What is the primary language spoken there?")
]

response_with_history = llm.invoke(messages)
print(response_with_history.content)
```
